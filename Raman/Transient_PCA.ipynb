```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd
import os

# --- Example Data (simulated spectra) ---
# Suppose 50 spectra, each with 200 wavelength points
cols = np.linspace(0,3,94)
rows = np.linspace(99, 1871, 254)
data = pd.read_csv('Averaged_test_data_40_20.csv', header=None)
data.index = rows
data.columns = cols
data = data[data.index <= 1200]
wavelengths = np.linspace(data.index[-1], data.index[0], len(data.index))  # nm, for example
spectra = np.array(data.T)


# --- Step 1: Standardize data (important for PCA) ---
scaler = StandardScaler(with_mean=True, with_std=True)
spectra_scaled = scaler.fit_transform(spectra)

# --- Step 2: Apply PCA ---
n_components = 30
pca = PCA(n_components=n_components)  # keep first 5 PCs
scores = pca.fit_transform(spectra_scaled)
loadings = pca.components_
explained_var = pca.explained_variance_ratio_

# --- Step 3: Plot results ---
plt.figure(figsize=(12, 8))

# Original spectra
plt.subplot(2, 2, 1)
plt.plot(wavelengths, spectra.T, alpha=0.5)
plt.title("Original Spectra")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Intensity")

# PCA loadings (principal components)
plt.subplot(2, 2, 2)
for i in range(3):
    plt.plot(wavelengths, loadings[i], label=f"PC{i+1} ({explained_var[i]*100:.1f}%)")
plt.title("PCA Loadings")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Loading weight")
plt.legend()

# PCA scores (sample projections)
plt.subplot(2, 2, 3)
plt.scatter(scores[:, 0], scores[:, 1], c='b', alpha=0.7)
plt.title("PCA Scores")
plt.xlabel("PC1")
plt.ylabel("PC2")

# Variance explained
plt.subplot(2, 2, 4)
plt.bar(range(1, n_components+1), explained_var*100)
plt.title("Explained Variance")
plt.xlabel("Principal Component")
plt.ylabel("Variance (%)")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import TruncatedSVD
from chemometrics.mcr import McrAR
import chemometrics.mcr.constraint as constraint

# --- Load your data (adapt if your file layout differs) ---
data = pd.read_csv('Averaged_test_data_40_20.csv', header=None)
#data = data[::-1]
rows = np.linspace( 1871, 99, 254)
cols = np.linspace(0, 3, 94)

data.index = rows
data.columns = cols
data = data[(data.index <= 850) & (data.index >=300)]


# D must be shape (n_samples, n_wavelengths)
D = data.T.values   # samples x wavelengths

# --- Choose number of components ---
n_components = 2   # pick the expected number of pure components

# --- Make a sensible initial estimate for C (concentrations) via truncated SVD ---
c_init = np.abs(TruncatedSVD(n_components=n_components).fit_transform(D))
# scale so values are roughly 0..1 (optional but helps)
c_init /= np.max(c_init)

# --- Instantiate MCR with constraints (no add_constraint method) ---
regression_type = 'OLS'
mcr = McrAR(
    c_regr= regression_type,
    c_constraints=[constraint.Nonneg(), constraint.Normalizer()],  # concentrations nonneg & normalized 
    st_regr= regression_type,
    st_constraints=[constraint.Nonneg()],                         # spectra nonneg
    max_iter=500,
    tol_increase=0.01
)

# --- Fit MCR (supply initial C) ---
mcr.fit(D, C=c_init)

# Results: fitted concentration matrix and spectra
C_est = mcr.C_      # shape (n_samples, n_components)
S_est = mcr.ST_.T   # chemometrics stores ST_ (components x wavelengths), .T gives (wavelengths x components)


#plt.plot(rows, data.T.iloc[49])



# --- Plot recovered spectra (components) ---
wavenumbers = data.index.values
plt.figure(figsize=(8,4))
for k in range(n_components):
    plt.plot(wavenumbers, mcr.ST_.T[:, k], label=f'Comp {k+1}')
#plt.gca().invert_xaxis()  # optional for Raman/IR style x-axis
plt.legend()
plt.xlabel('Raman Shift (cm-1)')
plt.ylabel('Intensity (AU)')
plt.title('Recovered component spectra (chemometrics McrAR)')
plt.show()

# --- Plot concentration profiles ---
plt.figure(figsize=(8,4))
for k in range(n_components):
    plt.plot(C_est[:, k], label=f'Comp {k+1}')
plt.legend()
plt.xlabel('Sample index')
plt.ylabel('Relative contribution')
plt.title('Estimated concentration profiles')
plt.show()
print(data.shape)
```
